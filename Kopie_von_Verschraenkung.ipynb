```python
print("import Libraries ....")
import numpy as np
import matplotlib.pyplot as plt
import random
from qiskit import QuantumCircuit
from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector

print("done.")
```

```python
# Simulator für Quantenschaltkreise -- Sampler --
def simulate(qc):
    print("***** Simulation ********")
    simulator = AerSimulator()
    circ = transpile(qc, simulator)
    # Run and get counts
    result = simulator.run(circ,shots=1000).result()
    counts = result.get_counts()
    print("Ergebnis der Simulation:")
    print(counts)
    # Ergebnis Histogramm
    display(plot_histogram(counts))
```

# Verschränkung

```python
# Define Quantumcircuit
qc1 = QuantumCircuit(2,1)
qc1.h(0)
qc1.cx(0,1)
qc1.measure(0,0)
qc1.draw("mpl")
```

```python
simulate(qc1)
```

```python
# Define Quantumcircuit
qc2 = QuantumCircuit(2,1)
qc2.h(0)
qc2.cx(0,1)
qc2.measure(1,0)
qc2.draw("mpl")
```

```python
simulate(qc2)
```

```python
# Define Quantumcircuit
qc3 = QuantumCircuit(2,2)
qc3.h(0)
qc3.cx(0,1)
qc3.barrier()
qc3.measure([0,1],[0,1])
qc3.draw("mpl")
```

```python
simulate(qc3)
```
